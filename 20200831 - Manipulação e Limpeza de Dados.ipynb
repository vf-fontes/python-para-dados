### Função que substitui automaticamente valores insconsistentes com valores semelhantes da própria coluna

```
def substitui_semelhantes(df, coluna, texto, semelhança = 47):
    """substitui as linhas de uma determinada coluna onde o valor possui semelhança maior que o informado"""
    # lista de valores únicos
    strings = df[coluna].unique()
    
    # 10 valores mais semelhantes ao texto
    parecidos = fuzzywuzzy.process.extract(texto, strings, 
                                         limit=10, scorer=fuzzywuzzy.fuzz.token_sort_ratio)
 
    # seleciona apenas os parecidos com semelhança > "semelhança"
    mais_parecidos = [parecidos[0] for parecidos in parecidos if parecidos[1] >= semelhança]
 
    # pega a linha dos mais_parecidos
    linha_parecidos = df[coluna].isin(mais_parecidos)
 
    # substitui pelo valor do texto 
    df.loc[linha_parecidos, coluna] = texto
    
    print("Substituição realizada.")
```

```
 # cria um array que começa no menor valor t ermina no maior valor, e cria 4 elementos dentro deste lin_space
categorias = np.linspace(df_terremoto.Magnitude.min(), df_terremoto.Magnitude.max(), 4)

# sempre quantidade do meu grupo -1
nomes_categorias = ["Fraco", "Médio", "Forte"]

# Aplica a Categorização
df_terremoto["Magnitude_Type"] = pd.cut(df_terremoto.Magnitude, categorias, labels=nomes_categorias, include_lowest=True)
df_terremoto.head()
```
