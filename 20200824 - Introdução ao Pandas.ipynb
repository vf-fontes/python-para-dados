### Mapeando valores:

```
media = df_casas['rent amount (R$)'].mean()
media
```

```
# forma mais longa de fazer o antetior
def zeraMedia(linha):
  linha['rent amount (R$)']=linha['rent amount (R$)']-media
  return linha

new_df_casas = df_casas.apply(zeraMedia, axis=1)
new_df_casas
```

```
#4.Estou criando um site para facilitar inquilinos a encontrar o melhor imóvel para alugar, um dos filtros que desejo criar no meu site é a filtragem por tamanho do imóvel: pequeno (até 100 m2), médio (de 100 m2 até 300 m2), grande (de 300 m2 até 1.000 m2) e muito grande (acima de 1.000 m2), implemente esse filtro.
# ==> PROFESSOR

def classifica(linha):
  if linha.area <=100:
    linha['classification'] = "Pequeno"
  elif 100<linha.area<=300:
    linha['classification'] = "Médio"
  elif 300<linha.area<=1000:
    linha['classification'] = "Grande"
  else:
    linha['classification'] = "Muito Grande"
  return linha

df_casas.apply(classifica, axis = 1)
```
