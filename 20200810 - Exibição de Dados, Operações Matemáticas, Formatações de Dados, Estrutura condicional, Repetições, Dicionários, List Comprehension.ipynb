## Cálculo de Raiz de um polinômio do segundo grau

$$p(x)=ax^2+bx+c$$

```
a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
c = float(input("Digite o valor de c: "))

delta=b**2-(4*a*c)

if delta > 0:
  x1=(-b+delta**(1/2))/(2*a)
  x2=(-b-delta**(1/2))/(2*a)
  print(f"As raízes do polinômio x^{a:.2f} + {b:.2f}x + {c:.2f} é {x1:.2f} e {x2:.2f}")
elif delta == 0:
  x1=-b/(2*a)
  x2=-b/(2*a)
  print(f"As raízes do polinômio x^{a:.2f} + {b:.2f}x + {c:.2f} é {x1:.2f} e {x2:.2f}")
else:
  print("Não existe raiz!")
```

##Cálcular a média de N Valores:

```
N = int(input("Digite a quantidade de valores: "))
L = []

i = 0
while i < N:  
  x = float(input("Digite o elemento %i: " %(i+1)))  
  L.append(x)  
  i += 1

M = sum(L)/len(L)
print(f"A média dos valores digitados: {M:.2f}")
```

Como criar uma lista com o quadrado dos valores da lista x?

```
x_2 = []
for i in x:
  x_2.append(i**2)
x_2
```

```
x_2 = [i*i for i in x]
x_2
```

```
x_2 = [i+i for i in x]
x_2
```

Como criar uma lista com quadrado dos valores pares da lista x?

```
x_3 = []
print(x)
for i in x:
  if i%2 ==0:
    x_3.append(i**2)
  else:
    x_3.append(i)

x_3
```

```
x_3 = [i*i for i in x if i % 2 ==0]
x_3
```

```
x_4 = [i*i if i%2==0 else i for i in x]
x_4
```

```
x_5 = ["pin" if i%4==0 else i for i in list(range(1,100))]
x_5
```

Exemplo de Webscraping

```
import webbrowser, sys, requests, bs4, json
CEP = input("Digite o CEP desejado: ")
res = requests.get('https://viacep.com.br/ws/'+CEP+'/json/')
soup = bs4.BeautifulSoup(res.text, 'html.parser')
JSON_Datalist = soup.get_text()
dados = json.loads(JSON_Datalist)
if 'erro' in dados:
  print("CEP inválido.")
else:
  print('CEP: %s' % dados['cep'])
  print('Endereço: %s' % dados['logradouro'])
  print('Complemento: %s' % dados['complemento'])
  print('Bairro: %s' % dados['bairro'])
  print('Cidade: %s' % dados['localidade'])
  print('Estado: %s' % dados['uf'])
```
