### Sort (Ordenação)

```
#ordem crescente
df_casas.sort_values(by = "rooms", ascending=True) 
```

```
#ordem decrescente
df_casas.sort_values(by = "rooms", ascending=False) 
```

```
#posso ordernar por quantos atributos eu quiser
df_casas.sort_values(by = ["rooms","city"]) 
```

```
# Agrupamento com ordenação
df_casas.groupby(["rooms","city"]).size().sort_values()
```

```
# 
df_casas.groupby(['rooms','city']).area.agg([len,min,max])
```

```
df_casas.groupby(['rooms','city']).area.agg([len,min,max]).sort_values(by = ['min','max'], ascending=False)
```

```
new_df_casas = df_casas.groupby(by = ['rooms','city']).area.agg(len)
new_df_casas
```
